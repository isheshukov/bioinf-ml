# 1. (6) Реализуйте алгоритм Gaussian naive Bayes для датасета spam.csv.

```python
class GaussianNB(object):
    def __init__(self):
        self.class_probs = None
        self.mean = None
        self.var = None
        self.class_labels = None

    def fit(self, data):
        """
        Sets:    [p_1, ..., p_n] -- class probabilities
                 [[m_1, m_2, ..., m_n],
                  ...,
                  ... ]] -- mean for each feature of each class

                 [[v_1, v_2, ..., v_n],
                  ...,
                  ... ]] -- variance for each feature of each class
        """

        self.class_labels = np.sort(np.unique(data.y))
        self.class_probs = np.bincount(np.sort(data.y))/data.y.size
        self.mean = np.array([np.mean(data.X[data.y == label], axis=0) for label in self.class_labels])
        self.var  = np.array([np.var (data.X[data.y == label], axis=0) for label in self.class_labels])
        
    def predict(self, x):
        """
        Computes \argmax P(y)*P(x_1|y)*...*P(x_n|y) over y \in Y
        """
        
        def cond_probs(x):
            """
               Input: [x_1, ... , x_n] -- array of features
               Output: [[p_1^1, ..., p_n^1],
                        ...,
                        [p_1^k, ..., p_n^k]], where k - number of classes, n - number of features
            """
            n = self.class_labels.size
            return np.exp(-np.power(np.tile(x, (n,1)) - self.mean, 2) / (2*self.var)) / np.sqrt(2*np.pi*self.var)
            
        return self.class_labels[np.argmax(np.product(np.c_[self.class_probs, cond_probs(x)], axis = 1))]
```

```python
nb = GaussianNB()
nb.fit(spam.train)
predict_y = np.array([nb.predict(x) for x in spam.val.X])
```

```python
print("Accuracy = {}".format(accuracy_score(spam.val.y, predict_y)))
fpr, tpr, _ = roc_curve(spam.val.y, predict_y)
plt.plot(fpr, tpr)
plt.title("Spam")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.show()
```

# 2. (6) Реализуйте алгоритм Multinomial naive Bayes для количества различных слов в датасете smsspam.csv. Т.е., если у вас в словаре N слов, то у каждой точки будет N признаков. Stemming крайне рекомендуем.

```python
class MultinomialNB(object):
    def __init__(self):
        self.class_probs = None
        self.class_labels = None
        
        self._X = None
        self._y = None
        self._p = None
        self._stemmer = PorterStemmer()
        self._label_mapping = None
        self._bin_mapping = None
    
    def fit(self, data):
        
        self.class_labels = np.unique(data.y)
        self._label_mapping = dict(enumerate(self.class_labels))
        self._bin_mapping = {y:x for x, y in self._label_mapping.items()}
        self._y = np.array([self._bin_mapping[i] for i in data.y])
        self.class_probs = np.bincount(np.sort(self._y))/self._y.size

        # Creating a dictionary
        rows = {}
        for i, msg in enumerate(data.X):
            tokenized = word_tokenize(msg[0])
            stemmed = [self._stemmer.stem(word.lower()) for word in tokenized]
            c = Counter(stemmed)
            rows.update({i: c})
        
        self._X = pd.DataFrame.from_dict(rows, orient='index').fillna(0).astype(int)          
        self._p = np.array([np.sum(self._X[data.y == label].values, axis=0) / (data.y == label).sum() for i, label in enumerate(self.class_labels)])
        
    def predict(self, x):
        """
        Computes \argmax P(y)*P(x_1|y)*...*P(x_n|y) over y \in Y
        """
        def to_histogram(x):
            tokenized = word_tokenize(x)
            stemmed = [self._stemmer.stem(word.lower()) for word in tokenized]
            c = Counter(stemmed)
            return np.array([c.get(i) if i in c else 0 for i in self._X.columns.values])
        
        def cond_probs(x):
            h = to_histogram(x)
            return factorial(np.sum(h)) / np.product(factorial(h)) * np.product(np.power(self._p, h), axis=1)
         
        return self.class_labels[np.argmax(np.product(np.c_[self.class_probs, cond_probs(x)], axis = 1))]
```

```python
nm = MultinomialNB()
nm.fit(smsspam.train)
```

```python
predict_y = np.array([nm.predict(x[0]) for x in smsspam.val.X])
```

```python
print("Accuracy = {}".format(accuracy_score(smsspam.val.y, predict_y)))
fpr, tpr, _ = roc_curve([nm._bin_mapping[i] for i in smsspam.val.y], [nm._bin_mapping[i] for i in predict_y])
plt.plot(fpr, tpr)
plt.title("SMS Spam")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.show()
```

# 3. (3) Реализуйте алгоритм линейной регрессии, и полиномиальной регрессии (для датасета noisysine – степеней от 2 до 5, для датасета hydrodynamics – степени 2) без регуляризации.

```python
class LinearRegression():
    def __init__(self, polynomial_degree=None):
        self._X = None
        self._y = None
        self._w = None
        self._pd = polynomial_degree
    
    def __str__(self):
        return "X = " + np.array2string(self._X) + \
               "\n y = " + np.array2string(self._y)
    
    def _transform_x(self, x):
        if self._pd is None:
            return np.c_[np.ones((x.shape[0],1)), x]
        
        if x.ndim == 1:
            X = np.c_[np.ones((x.shape[0],1)), x]
            for i in range(2, self._pd+1):
                X = np.c_[X, np.power(x, i)]
        else:
            if self._pd == 2:
                X = np.c_[np.ones((x.shape[0],1)), x]
                for i, j in combinations_with_replacement(range(x.shape[1]), 2):
                    X = np.c_[x, x[:, i]*x[:, j]]
            else:
                raise ValueError("Dim > 1, degree > 2 not supported")
            
        return X
            
    def fit(self, data):
        self._y = data.y
        self._X = self._transform_x(data.X)
    
        self._w = (np.linalg.pinv(self._X) @ self._y)
        
    def predict(self, x):
        try:
            return (self._transform_x(x) * self._w).sum(1)
        except:
            raise ValueError("{} is not an numpy array".format(x))
```

```python
lr = LinearRegression()
lr.fit(noisysine.train)

plt.scatter(*noisysine.val)

x = np.linspace(-3, 40, 2)
plt.plot(x, lr.predict(x), 'r')
plt.ylim(-2,2)
plt.title("Noisy sine (validate), linear regression")
plt.show()
```

```python
print("Sine R^2 score = {}".format(r2_score(noisysine.val.y, lr.predict(noisysine.val.X))))
```

```python
class RidgeRegression(LinearRegression):
    def __init__(self, polynomial_degree=None, alpha=0):
        self._X = None
        self._y = None
        self._w = None
        self._pd = polynomial_degree
        self.alpha = alpha

    def fit(self, data):
        self._y = data.y
        self._X = self._transform_x(data.X)
    
        n = (self._X.T @ self._X).shape[0]
        self._w = np.linalg.inv((self._X.T @ self._X + self.alpha*np.identity(n))) @ self._X.T @ self._y
```
