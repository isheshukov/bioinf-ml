```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools as it
import random
from typing import List
from tqdm import tqdm_notebook
from scipy.spatial.distance import pdist, squareform
from scipy.spatial import cKDTree
from itertools import product
```

# (2) Реализуйте Монте-Карло поиск. 

```python
data = pd.read_csv("../data/tsp.csv", index_col=0, header=None).values
```

```python
plt.figure(figsize=(8,8))
plt.plot(data, 'b.')
plt.show()
```

```python
distances = squareform(pdist(data, metric='cityblock'))
```

```python
def eval_path(path):
    s = 0
    for i in range(len(path)-1):
        s += distances[i][i+1]
    return s

def random_path(distances):
    a = list(range(len(data)))
    np.random.shuffle(a)
        
    return data[a], eval_path(a)
```

```python
min_path, min_dist = min((random_path(distances) for _ in tqdm_notebook(range(10000000))), key = lambda t: t[1])
```

```python
def plot_path_l1(path):
    for i in range(len(path)-1):
        # (x1, y1) -> (x2, y1) -> (x2,y2)      
        plt.plot([path[i][0], path[i+1][0]], [path[i][1], path[i][1]], 'r-')
        plt.plot([path[i+1][0], path[i+1][0]], [path[i][1], path[i+1][1]], 'r-')
    
    plt.plot(path[0][0], path[0][1], 'go') # start
    plt.plot(path[-1][0], path[-1][1], 'bo') # end
```

```python
plt.figure(figsize=(20,15))
plt.scatter(data[:,0], data[:,1])

plot_path_l1(min_path)
print(f"Minimal distance {min_dist}")
plt.show()
```

# (2) Реализуйте поиск случайным блужданием.

```python
a = list(range(len(data)))
def random_walk():
    best_walk = None
    best_dist = np.inf
    for _ in tqdm_notebook(range(10000000)):
        l, r = np.random.randint(0, len(data), size=2)
        a[l], a[r] = a[r], a[l]
        
        s = 0
        for i in range(len(a)-1):
            s += distances[i][i+1]
        
        if s < best_dist:
            best_dist = s
            best_walk = a

    return data[best_walk], best_dist
```

```python
best_walk, best_dist = random_walk()
```

```python
plt.figure(figsize=(20,15))
plt.scatter(data[:,0], data[:,1])

plot_path_l1(best_walk)
print(f"Minimal distance {best_dist}")
plt.show()
```

# (4) Реализуйте Hill Climb.


```python
def reversed_sections(walk):
    '''generator to return all possible variations 
      where the section between two cities are swapped'''
    for i,j in product(range(len(walk)), repeat=2):
        if i != j:
            copy=walk[:]
            if i < j:
                copy[i:j+1]=np.flip(walk[i:j+1])
            else:
                copy[i+1:]=np.flip(walk[:j])
                copy[:j]=np.flip(walk[i+1:])
            if copy != walk: # no point returning the same tour
                yield copy
```

```python
def hillclimb(move_operator=reversed_sections, max_evaluations = 1000000):
    
    best_walk=list(range(len(data)))
    best_dist=eval_path(best_walk)
    
    num_evaluations=1
    
    while num_evaluations < max_evaluations:
        move_made=False
        for next in move_operator(best_walk):
            if num_evaluations >= max_evaluations:
                break
            
            next_score=eval_path(next)
            num_evaluations+=1
            if next_score > best_dist:
                best=next
                best_dist=next_dist
                move_made=True
                break
            
        if not move_made:
            break
    
    return best_dist, best_walk
```

```python
best_dist, best_walk = hillclimb()
```

```python
plt.figure(figsize=(20,15))
plt.scatter(data[:,0], data[:,1])

plot_path_l1(data[best_walk])
print(f"Minimal distance {best_dist}")
plt.show()
```
